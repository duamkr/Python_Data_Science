## 스타일 지정

```python
plt.title("'rs--'스타일의 plot'")
plt.plot([10,20,30,40],[1,4,9,16], 'rs--')     # r = red,  s = marker종류, -- = dashed line style
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=8, mec="g", mew=3, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

## 그림 범위 지정(xlim, ylim)

```python
plt.title('x축, y축의 범위 설정')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=8, mec="g", mew=3, mfc="r")
plt.xlim(0,50)     # x축 범위 지정
plt.ylim(-10,30)  # y축 범위 지정
plt.show()
```

## 틱 설정(xticks, yticks)

```python
X = np.linspace(-np.pi, np.pi, 256)
```

```python
C = np.cos(X)
```

```python
plt.title("x축과 y축의 tick label 설정")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
# 틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자식을 넣을 수도 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

## 그리드 설정(grid)

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 제거")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(False)     # 표시 = True
plt.show()
```

## 여러개의 선을 그리기(plot(x데이터, y데이터,스타일, x1데이터, y1데이터, 스타일 -- 반복)

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

## 홀드명령

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
# plt.hold(True)   # <- 1,5 버전에서는 이 코드가 필요하다.
plt.plot([9, 16, 4, 1],
         c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")
# plt.hold(False)  # <- 1,5 버전에서는 이 코드가 필요하다.
plt.show()
```

## 범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)     # 2 = upper left
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=3)     # 2 = lower left
plt.show()
```

## x축, y축 라벨, 타이틀(xlabel,ylabel)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

#### 연습문제 4.4.1
##### 여러가지 함수를 사용하여 아래 조건에 맞는 그래프를 그린다.
- xlabel, ylabel, title을 모두 갖추고 있어야 한다.
- 하나의 Figure(일단, 그림이라고 이해한다. 아래에 자세한 설명이 있다.)에 3개 - 이상의 Plot을 그린다.
- 각 Plot은 다른 선, 마크, 색 스타일을 가진다.
- legend는 그래프와 겹치지 않는 곳에 위치 시키도록 한다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S, Z = np.cos(X), np.sin(X), np.tan(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, c = "r" ,marker = 's',ls="--", label="cosine")
plt.plot(X, S, c = "g" ,ls=":", marker = 'o', label="sine")
plt.plot(X, Z, c = "b" ,ls=":", marker = 'v', label="tangent")
plt.legend(loc=2)
plt.show()
```
